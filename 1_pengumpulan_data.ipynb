## 1.2 Import Library

```python
import nest_asyncio
import twint
import pandas as pd

nest_asyncio.apply()# used once to enable concurrent actions within a Jupyter notebook.
```

## 1.3 Proses Crawling

```python
# configure
c = twint.Config()
c.Limit = 5000

# bahasa tweet
c.Lang = "id"

# menyembunyikan output hasil scrap
c.Hide_output = True

# scrap pada tanggal - sampai
c.Since = '2022-02-10'
c.until= '2022-03-10'

# scrap berdasarkan hashtag / key
c.Search = ['logo halal'] # jika berdasarkan hashtag tak perlu pake list langsung '#hashtag'

# simpan ke pandas csv
c.Pandas = True
c.Store_csv = True

# mulai scrap
twint.run.Search(c)

# pandas dataframe
df = twint.storage.panda.Tweets_df

# save ke google drive
df.to_csv('/content/drive/MyDrive/Dataset/tugas-akhir/original/logo_halal_tweet.csv')
```
